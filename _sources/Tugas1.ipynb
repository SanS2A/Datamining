```python
#import library
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
url = "https://raw.githubusercontent.com/SanS2A/dataset/main/heart.csv"

c = pd.read_csv(url)
c
```

## **Kolom/Fitur/Atribut**



- age
- sex
- chest pain type (4 values)
- resting blood pressure
- serum cholestoral in mg/dl
- fasting blood sugar > 120 mg/dl
- resting electrocardiographic results (values 0,1,2)
- maximum heart rate achieved
- exercise induced angina
- oldpeak = ST depression induced by exercise relative to rest
- the slope of the peak exercise ST segment
- number of major vessels (0-3) colored by flourosopy
- thal: 0 = normal; 1 = fixed defect; 2 = reversable defect

##2.   Membuat Statistik Deskriptif dari data (setiap atribut yang bertipe numerik)


Statistik Deskriptif yang ada:
* count - Jumlah nilai yang tidak kosong.
* mean - Nilai rata-rata (rata-rata).
* std - Standar deviasi.
* min - nilai minimum.
* 25% - Persentil 25%*.
* 50% - Persentil 50%*.
* 75% - Persentil 75%*. 
* maks - nilai maksimum.
* median - nilai tengah.
* mode- data yang paling banyak muncul



```python
#menampilkan informasi dari kolom data numerik 
c.describe()
```

## Median -> Nilai tengah

```python
med = {'age':c['age'].median(),
		'sex': c['sex'].median(),'cp':c['cp'].median(),
    'trestbps':c['trestbps'].median(),'chol':c['chol'].median(),
    'fbs':c['fbs'].median(),'restecg':c['restecg'].median(),
    'thalach':c['thalach'].median(),'exang':c['exang'].median(),
    'oldpeak':c['oldpeak'].median(),'slope':c['slope'].median(),
    'ca':c['ca'].median(),'thal':c['thal'].median(),
     'target':c['target'].median()
    }

df = pd.DataFrame(med, index=['median'])
df
```

```python
c["age"].median()
```

## Mean -> Rata-rata

```python
mea = {'age':c['age'].mean(),
		'sex': c['sex'].mean(),'cp':c['cp'].mean(),
    'trestbps':c['trestbps'].mean(),'chol':c['chol'].mean(),
    'fbs':c['fbs'].mean(),'restecg':c['restecg'].mean(),
    'thalach':c['thalach'].mean(),'exang':c['exang'].mean(),
    'oldpeak':c['oldpeak'].mean(),'slope':c['slope'].mean(),
    'ca':c['ca'].mean(),'thal':c['thal'].mean(),
     'target':c['target'].mean()
    }

df = pd.DataFrame(mea, index=['mean'])
df
```

```python
c["age"].mean()
```

## Modulus -> Nilai yang sering muncul

```python
mod = {'age':c['age'].mode(),
		'sex': c['sex'].mode(),'cp':c['cp'].mode(),
    'trestbps':c['trestbps'].mode(),'chol':c['chol'].mode(),
    'fbs':c['fbs'].mode(),'restecg':c['restecg'].mode(),
    'thalach':c['thalach'].mode(),'exang':c['exang'].mode(),
    'oldpeak':c['oldpeak'].mode(),'slope':c['slope'].mode(),
    'ca':c['ca'].mode(),'thal':c['thal'].mode(),
     'target':c['target'].mode()
    }

df = pd.DataFrame(mod, index=['mode'])
df
```

```python
c["age"].mode()
```

## 3.   Mendeskripsikan kemiringan/skewned dari data setiap variabel(bertipe numerik)



```python
sns.distplot(c['age'],kde=True,
             bins= int(150/5),color='lightblue',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Positive Skewned')
  
  
```

```python
sns.distplot(c['sex'],kde=True,
             bins= int(150/5),color='lightblue',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Symmetric Skewned')
  
```
